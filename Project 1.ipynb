```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from sympy.plotting import plot3d

x, y = symbols('x y')
```

```python
u = 5 * (x**2) - y**2
v = 4*y - (sin(x) + cos(y))
```

```python
f1 = lambdify([x,y], u)
f2 = lambdify([x,y], v)

n= np.arange(-4, 4, 0.01)

Xs, Ys = np.meshgrid(n, n)

xs = f1(Xs, Ys)
ys = f2(Xs, Ys)

plt.contour(Xs, Ys, xs ,0, colors="red", linewidths=1)
plt.contour(Xs, Ys, ys ,0, colors="blue",  linewidths=1)
plt.show()
```

```python
F = Matrix([u, v])
F.jacobian([x, y])
```

```python
p = Matrix([x, y])
G = p - F.jacobian([x, y]).inv() * F
G
```

```python
plot3d(G[1], (x, -2, 2), (y, -2, 2))
```

```python
plot3d(G[0], (x, 0.1, 0.5), (y, 0.15, 0.4))
```
